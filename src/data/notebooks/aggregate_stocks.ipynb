```python
import pandas as pd
import numpy as np
import time
import datetime
```

```python
df_stocks = pd.read_csv("data/tab2/Stocks.txt", sep=" ")
df_stocks
```

```python
df_stocks["year"] = df_stocks["year_week"].str[:4].astype(int)
df_stocks["week"] = df_stocks["year_week"].str[5:].astype(int)
```

```python
df_stocks["Year-Week"] = df_stocks["year_week"]
df_stocks.drop(df_stocks.loc[:,"country":"cumulative_count_deaths"].columns, axis=1, inplace=True)
```

```python
df_stocks
```

```python
var_names = ["Open","High","Low","Close","Volume","Adj Close"]
```

```python
df_combined = df_stocks.melt(id_vars=["year","week"])
df_combined
```

```python
df_test = df_combined.groupby(["year","week","variable"]).sum()
```

```python
df_test
```

```python
df_combined["Brand Abbreviation"] = df_combined["variable"].str.split(" - ").str[0]
df_combined["OHLC"] = df_combined["variable"].str.split(" - ").str[1]
```

```python
# Transpose the OHLC column
df_combined = df_combined.pivot_table(index=["year","week","Brand Abbreviation"], columns="OHLC", values="value").reset_index()
df_combined = df_combined.rename_axis(None, axis=1)
```

```python
df_combined.sort_values(by=["Brand Abbreviation","year","week"], inplace=True)
df_combined = df_combined.round(2)
```

```python
df_combined
```

```python
df_ABNB = pd.read_csv("data/tab2/ABNB_stock_data.csv")
df_ABNB = df_ABNB.round(2)
```

```python
df_ABNB["year"] = df_ABNB["Date"].str[:4].astype(int)
df_ABNB["Brand Abbreviation"] = "ABNB"
df_ABNB.rename(columns={"Adj Close":"Adj.Close"}, inplace=True)
df_ABNB
```

```python
# Get week number from date
df_ABNB["week"] = pd.to_datetime(df_ABNB["Date"]).dt.week
```

```python
# Add df_ABNB to df_combined
df_combined = pd.concat([df_combined, df_ABNB[["year","week","Brand Abbreviation","Adj.Close","Close","High","Low","Open","Volume"]]], axis=0)
```

```python
df_combined
```

```python
df_combined["Brand Abbreviation"].unique()
```

```python
# Create a dictionary to map the brand abbreviations to the full brand names and create new column
brand_names = {"AAPL":"Apple","ABNB":"Airbnb","HNNMY": "HM", "KO":"Coca-Cola", "NKE":"Nike", "SBUX":"Starbucks","PEP": "Pepsi"}
df_combined["Brand Name"] = df_combined["Brand Abbreviation"].map(brand_names)
df_combined["Year-Week"] = df_combined["year"].astype(str) + "-" + df_combined["week"].astype(str).str.zfill(2)
df_combined
```

```python
df_combined[df_combined["Brand Name"] == "Apple"]
```

```python
df_combined["week increment"] = 0
for brand in df_combined["Brand Name"].unique():
    if brand != "Airbnb":
        df_combined.loc[df_combined["Brand Name"] == brand, "week increment"] = np.arange(0,df_combined[df_combined["Brand Name"] == brand].shape[0]) + 1
```

```python
df_combined
```

```python
df_combined[(df_combined["Brand Name"] == "Apple") & (df_combined["week increment"] >= 76)]["week increment"].to_list()
```

```python
# Set Airbnb week equal to other brands
df_combined.loc[df_combined["Brand Name"] == "Airbnb", "week increment"] = df_combined[(df_combined["Brand Name"] == "Apple") & (df_combined["week increment"] >= 76)]["week increment"].to_list()
```

```python
df_combined["week increment"] = df_combined["week increment"].astype(int)
df_combined
```

```python
# Create lagged variable for Adj.Close
df_combined["Adj.Close_Lag_1"] = df_combined.groupby(["Brand Name"])["Adj.Close"].shift(1).to_frame()
```

```python
df_combined["pct_change"] = (df_combined["Adj.Close"] - df_combined["Adj.Close_Lag_1"]) / df_combined["Adj.Close_Lag_1"]
df_combined
```

```python
df_combined.loc[df_combined["year"] == 2021, "week"] = df_combined.loc[df_combined["year"] == 2021, "week"] + 1
dates = df_combined.year*100+df_combined.week - 1
df_combined['Date'] = pd.to_datetime(dates.astype(str) + '0', format='%Y%W%w')
df_combined.loc[df_combined["year"] == 2021, "week"] = df_combined.loc[df_combined["year"] == 2021, "week"] - 1
df_combined
```

```python
df_combined = pd.read_csv("data/tab2/aggregated_stocks_values.csv")
```

```python
brands = df_combined["Brand Abbreviation"].unique()
```

```python
# Scrape stock data from yahoo finance
df_day = pd.DataFrame()
df_week = pd.DataFrame()
for brand in brands:
    period1 = int(time.mktime(datetime.datetime(2019, 7, 1, 00, 00).timetuple()))
    period2 = int(time.mktime(datetime.datetime(2021, 3, 28, 23, 59).timetuple()))
    if brand == "NKE":
        period2 = int(time.mktime(datetime.datetime(2021, 4, 25, 23, 59).timetuple()))
    day = "1d"
    week = "1wk"
    
    query_string_day = f'https://query1.finance.yahoo.com/v7/finance/download/{brand}?period1={period1}&period2={period2}&interval={day}&events=history&includeAdjustedClose=true'
    query_string_week = f'https://query1.finance.yahoo.com/v7/finance/download/{brand}?period1={period1}&period2={period2}&interval={week}&events=history&includeAdjustedClose=true'
    df_day_brand = pd.read_csv(query_string_day).round(2)
    df_week_brand = pd.read_csv(query_string_week).round(2)
    
    df_day_brand["Brand Abbreviation"] = brand
    df_week_brand["Brand Abbreviation"] = brand
    df_day = pd.concat([df_day, df_day_brand], axis=0)
    df_week = pd.concat([df_week, df_week_brand], axis=0)

# ticker = 'AAPL'
# period1 = int(time.mktime(datetime.datetime(2020, 12, 1, 23, 59).timetuple()))
# period2 = int(time.mktime(datetime.datetime(2020, 12, 31, 23, 59).timetuple()))
# interval = '1d' # 1d, 1m

# query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

# df = pd.read_csv(query_string).round(2)
# df
# print(df)
# df.to_csv('AAPL.csv')
```

```python
df_day["pct_change"] = df_day.groupby(["Brand Abbreviation"])["Close"].pct_change(1)*100
df_week["pct_change"] = df_week.groupby(["Brand Abbreviation"])["Close"].pct_change(1)*100
df_day["Brand Name"] = df_day["Brand Abbreviation"].map(brand_names)
df_week["Brand Name"] = df_week["Brand Abbreviation"].map(brand_names)
```

```python
df_day.to_csv("data/tab2/aggregated_stocks_values_days.csv", index=False)
df_week.to_csv("data/tab2/aggregated_stocks_values_weeks.csv", index=False)
```
