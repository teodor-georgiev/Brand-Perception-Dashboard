```python
import pandas as pd
```

```python
df_yougov = pd.read_csv("data/tab2/YouGov_Gesamt.csv", sep=";")
```

```python
# create empty column with name Date
df_yougov["Date"] = ""
```

```python
# start date 2019-07-01 end date 2021-02-28
# fill the date column for every brand with the dates for every day between the start date and end date
for brand in df_yougov["Brand"].unique():
    df_yougov.loc[df_yougov["Brand"] == brand, "Date"] = pd.date_range(
        start="2019-07-01", end="2021-02-28", freq = "D"
    )
```

```python
# format yyyy-mm-dd
df_yougov["Date"] = df_yougov["Date"].astype(str).str[:10]
df_yougov
```

```python
# create a dictionary and repeat every column as a key and respected value
# example: "Brand": "Brand", "Date": "Date", "YouGov": "YouGov"
# df_dict ={ col: col for col in df_yougov.columns[1:-1]}
# df_dict
```

```python
brand_index = ["Impression", "Quality", "Value", "Reputation", "Satisfaction", "Recommend"]
```

```python
df_yougov["BrandIndex"] = df_yougov[brand_index].mean(axis=1).round(2)
```

```python
# Change the position of the column Date to be last
cols = list(df_yougov.columns)
cols = cols[:-2]
cols = cols + ["BrandIndex", "Date"]
df_yougov = df_yougov[cols]
df_yougov
```

```python
df_yougov.to_csv("data/tab2/yougov_daily.csv", index=False)
```

```python
df_stocks_weeks = pd.read_csv("data/tab2/aggregated_stocks_values_weeks.csv")
```

```python
df_yougov["Date"] = pd.to_datetime(df_yougov["Date"])
df_yougov["week"] = df_yougov["Date"].dt.isocalendar().week
df_yougov["year"] = df_yougov["Date"].dt.year
df_yougov["week_year"] = df_yougov["year"].astype(str) + "-" + df_yougov["week"].astype(str)
df_yougov["week_year"].replace({"2019-1": "2020-1","2021-53" : "2020-53"}, inplace=True)
```

```python
df_stocks_weeks["Date_64"] = pd.to_datetime(df_stocks_weeks["Date"])
df_stocks_weeks["week"] = df_stocks_weeks["Date_64"].dt.isocalendar().week
df_stocks_weeks["year"] = df_stocks_weeks["Date_64"].dt.year
df_stocks_weeks["week_year"] = df_stocks_weeks["year"].astype(str) + "-" + df_stocks_weeks["week"].astype(str)
df_stocks_weeks["week_year"].replace({"2019-1": "2020-1", "2021-53" : "2020-53"}, inplace=True)
```

```python
# # Get week year from date
# df_yougov["Date"] = pd.to_datetime(df_yougov["Date"])
# df_yougov["week_year"] = df_yougov["Date"].dt.strftime("%Y-%U")
# df_stocks_weeks["Date_64"] = df_stocks_weeks["Date"].astype("datetime64")
# df_stocks_weeks["week_year"] = df_stocks_weeks["Date_64"].dt.strftime("%Y-%U")

# df_stocks_weeks["week_year"].replace({"2021-00": "2020-52"}, inplace=True)
# df_yougov["week_year"].replace({"2021-00": "2020-52"}, inplace=True)
```

```python
df_yougov["date_week"] = df_yougov["week_year"].replace(df_stocks_weeks["week_year"].values, df_stocks_weeks["Date"].values)
df_yougov_weekly = df_yougov.groupby(["Brand", "date_week"]).mean().reset_index()
```

```python
df_yougov_weekly = df_yougov_weekly.round(1)
df_yougov_weekly.rename(columns={"date_week":"Date"},inplace = True)
```

```python
df_yougov_weekly.drop(columns=["week", "year"], inplace=True)
```

```python
df_yougov = pd.read_csv("data/tab2/yougov_daily.csv")
```
