```python
import pandas as pd
import glob
import os
```

```python
df_test = pd.read_csv("hashtags/Airbnb_2019-07-01_2019-08-01.csv", sep=";")
```

```python
df_test
```

```python
path = r'hashtags' # use your path
files = glob.glob(os.path.join(path , "*.csv"))
files
```

```python
# brands = [str.split(str.rsplit(i, sep="\\", maxsplit=1)[1], sep="_", maxsplit=1)[0] for i in files if i not in brands]
brands_files = [str.split(str.rsplit(i, sep="\\", maxsplit=1)[1], sep="_", maxsplit=1)[0] for i in files]
brands = [*set(brands_files)]
brands
```

```python
df_aggregate = pd.DataFrame(columns=df_test.columns)
df_aggregate
```

```python
for i in files:
    df_file = pd.read_csv(i, sep=";")
    df_file["Unnamed: 0"] = str.split(str.rsplit(i, sep="\\", maxsplit=1)[1], sep="_", maxsplit=1)[0]
    df_aggregate = pd.concat([df_aggregate, df_file], axis=0)
```

```python
df_aggregate
```

```python
df_aggregate.rename(columns={"Unnamed: 0": "brand"}, inplace=True)
```

```python
df_analyze = df_aggregate[["brand","date","language","content","sentiment_neg",	"sentiment_neu"	,"sentiment_pos"]]
```

```python
df_analyze
```

```python
df_analyze.to_csv("hashtags/aggregated_data.csv", sep=",", index=False)
```
