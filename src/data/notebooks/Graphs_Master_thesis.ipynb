```python
import pandas as pd
# import kaleido
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df_stocks_days = pd.read_csv("../tab2/aggregated_stocks_values_days.csv")
```

```python
df_stocks_days = df_stocks_days[df_stocks_days["Brand Name"] != "Airbnb"]
df_stocks_days = df_stocks_days[df_stocks_days["Brand Name"] != "Pepsi"]
df_stocks_days
```

```python
df_stocks_days["cumulative_returns"] = df_stocks_days.groupby(["Brand Abbreviation"])["Close"].transform(lambda x: (x / x.iloc[0]-1)*100)
```

```python
df_stocks_weeks = pd.read_csv("../tab2/aggregated_stocks_values_weeks.csv")
df_stocks_weeks = df_stocks_weeks[df_stocks_weeks["Brand Name"] != "Airbnb"]
df_stocks_weeks = df_stocks_weeks[df_stocks_weeks["Brand Name"] != "Pepsi"]
df_stocks_weeks["cumulative_returns"] = df_stocks_weeks.groupby(["Brand Abbreviation"])["Close"].transform(lambda x: (x / x.iloc[0]-1)*100)
df_stocks_weeks.rename(columns={"Brand Name": "Brand"}, inplace=True)
df_stocks_weeks["Brand"] = df_stocks_weeks["Brand"].str.replace("HM", "H&M")
# sort by brand name
df_stocks_weeks.sort_values(by=["Brand","Date"], inplace=True)
df_stocks_weeks
```

```python
config = {
    'toImageButtonOptions': {
        'format': 'png', # one of png, svg, jpeg, webp
        'filename': 'custom_image',
        # 'height': 5000,
        # 'width': 700,
        'scale':4 # Multiply title/legend/axis/canvas sizes by this factor
    }
    }
```

```python
fig = px.line(df_stocks_weeks, x="Date", y="cumulative_returns", color="Brand",
              # symbol = "Brand",
              )
fig.update_layout(margin=dict(l=75, r=20, t=20, b=20),
                  template= "simple_white",height=500, width=900, xaxis_title="Date", yaxis_title="Cumulative Returns (%)",
                  legend=dict(
                  yanchor="top",
                  y=0.97,
                  xanchor="left",
                  x=0.02,
                  bordercolor="Black",
                  borderwidth=1,),
                  yaxis_title_standoff = 10,
                  font=dict(size=12, color="black"),
)
fig.update_xaxes(showline=True,  linecolor='black', mirror=True)
fig.update_yaxes(showline=True,  linecolor='black', mirror=True)
# Add custom markers for each brand
# fig.update_traces(mode="lines", line=dict(width=2.5, color="DarkSlateGrey",dash = "dot"), selector=dict(legendgroup="Nike"))
#  ['solid', 'dot', 'dash', 'longdash', 'dashdot', 'longdashdot']
fig.add_hline(y=0, line_width=1.5, line_dash="longdash", line_color="black",opacity=0.7)


fig.show(config=config)
```

### Weekly polarity boxplot

```python
df_tweets_sentiment_weekly = pd.read_csv("../tab2/twitter_sentiment_weekly_percent.csv")
df_tweets_sentiment_weekly = df_tweets_sentiment_weekly[df_tweets_sentiment_weekly["brand"] != "Airbnb"]
df_tweets_sentiment_weekly["brand"] = df_tweets_sentiment_weekly["brand"].str.replace("HM", "H&M")
```

### Overall Count of Tweets

```python
df_tweets_no_content = pd.read_csv("../tab2/preprocessed_tweets_no_content.csv")
df_tweets_no_content = df_tweets_no_content[df_tweets_no_content["brand"] != "Airbnb"]
df_tweets_no_content["brand"] = df_tweets_no_content["brand"].str.replace("HM", "H&M")
```

```python
overall_tweets_count = df_tweets_no_content[["brand","date"]].groupby("brand").count().reset_index()
overall_tweets_count.rename(columns={"date": "tweets_count"}, inplace=True)
```

```python
overall_tweets_count.sort_values(by="tweets_count", ascending=False, inplace=True)
```

```python
tweets_count_fig = px.bar(overall_tweets_count, x="brand", y="tweets_count",text_auto='.3s')
tweets_count_fig.update_layout(
    template= "simple_white",height=500, width=900, xaxis_title="Brand", yaxis_title="Number of tweets",font= dict(size=15),
    margin=dict(l=75, r=20, t=20, b=20)
)
tweets_count_fig.update_xaxes(showline=True,  linecolor='black', mirror=True)
tweets_count_fig.update_yaxes(showline=True,  linecolor='black', mirror=True)
tweets_count_fig.update_traces(marker_line_color='black', marker_line_width=1, opacity=1,width=0.4)
# change color of the bars
tweets_count_fig.update_traces(marker_color='#0072bd',textfont_size=12, textangle=0, textposition="outside", cliponaxis=False)
tweets_count_fig.show(config=config)
```

```python
df_yougov_daily = pd.read_csv("../tab2/yougov_daily.csv")
df_yougov_daily = df_yougov_daily[df_yougov_daily["Brand"] != "Airbnb"]
df_yougov_daily = df_yougov_daily[df_yougov_daily["Brand"] != "BMW"]
df_yougov_daily = df_yougov_daily[df_yougov_daily["Brand"] != "Pepsi"]
df_yougov_daily["Brand"] = df_yougov_daily["Brand"].str.replace("HM", "H&M")
df_yougov_daily["Date"] = pd.to_datetime(df_yougov_daily["Date"])
df_yougov_daily["Month"] = df_yougov_daily["Date"].dt.strftime("%Y-%m")
df_yougov_daily["Month"] = pd.to_datetime(df_yougov_daily["Month"])
df_yougov_daily["Month"] = df_yougov_daily["Month"] + pd.offsets.MonthEnd(0)
df_yougov_daily
```

```python
df_aggregate = pd.read_csv("../tab2/aggregated_stocktwits.csv")
```

```python
ignore_list = ["Airbnb", "Pepsi"]
df_aggregate["bullish"] = (df_aggregate["trend"] == "Bullish").astype(int)
df_aggregate["bearish"] = (df_aggregate["trend"] == "Bearish").astype(int)
df_aggregate = df_aggregate.query("brand not in @ignore_list")
df_aggregate
```

```python
df_sentiment_count = df_aggregate[["brand","bullish","bearish"]].groupby("brand").sum().reset_index()
df_sentiment_count["overall"] = df_stocktwits_count_overall["stocktwits_count"]
df_sentiment_count["sentiment_percentage"] = ((df_sentiment_count["bullish"] + df_sentiment_count["bearish"]) / df_sentiment_count["overall"] * 100).round(2)
df_sentiment_count = df_sentiment_count.sort_values(by="sentiment_percentage", ascending=False)
df_sentiment_count["brand"].replace({"HM":"H&M"}, inplace=True)
df_sentiment_count
```
